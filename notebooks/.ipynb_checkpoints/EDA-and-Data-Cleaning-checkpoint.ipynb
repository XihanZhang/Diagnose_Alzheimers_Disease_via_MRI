# Exploratory Data Analysis & Data Cleaning

## Recall the Research Question
Given the data sets suffer from a lot of **missing follow-up data**, we focused our main research question on building classifier models for patients **baseline diagnosis** using patients **brain imaging data from their first visits, adjusting for basic demographic information.**

## Exploration on ADNI Data Base
**(1)** The ADNI data base collected the data from **3 phases** of the longitudinal study: 
- ADNI1
- ADNIGO
- ADNI2

Each phase applied **different protocol**, and to make sure all subjects were undergone the same protocol, we **focused our project on ADNI1**, which has the **largest amount of participant**.

**(2)** ADNI1 includes **4 subsets of data**: 
- clinical data
- genetics data
- MRI data
- PET image data

We choose to **focus on MRI data**, because PET data has much less information than MRI data.


## How We Construct Our Data Sets

**Step1:** We started with the **adnimerge table** *(ADNIMERGE.csv)* which merges together several of the key variables from various case report forms and biomarker lab summaries across all the 3 ADNI protocols. We :
- Explored demographic distribution
- Compared data qualities among the 3 phases
- Tracked longitudinal scores 
- Plotted distribution of the 6 imaging variables. 

After all the basic exploration, **4 demographic features** and **6 imaging features** were left, which constitute the **base set**. We built our **base models** on this base set.

**Step2:** Then we **merged brain imaging features from UPENN Hierarchical Parcellation of MRI data** set *(UPENNROI_MARS_06_01_16.csv)* into the base set. We:

- Checked collinearity of the predictors
- Cleaned the hierarchical variables
- Combined symmetric features

Then we built our **final models** based on this **compositional set**.

#### (2) outcome selection

We selected **patients’ categorical diagnosis at baseline*(“DX_bl”)*** as the outcome variable, because it
has only 3 **categories *(CN, LMCI, AD)*** and is more reliable.

We also explored the **MMSE at baseline** as a surrogate outcome and may attempt to build up a model
for this variable into a model as well. The ideal case is to identify patients’ subgroups based on the
shape of their MMSE trajectories. Fixed effect longitudinal model might be attempted and individual
level coefficients extracted from the mixed effect longitudinal model will be utilized for patient
stratification. The trajectories of MMSE over time are generated using 35 randomly selected patients.
Due to the **imbalance and insufficient variation** in MMSE, we decided to **shelve the modeling of this
variable.** *(Figure 2)*

### (2) UPENN - Hierarchical Parcellation of MRI using multi-atlas labeling methods *(UPENNROI_MARS_06_01_16.csv)*

We use the brain imaging features from UPENN Hierarchical Parcellation of MRI data set (UPENN). UPENN dataset only contains **Baseline MRI scans of ADNI1 participants**, so its nature fits perfectly with our project question and the processed Adnimerge dataset above.  UPENN data set includes:

- 145 individual single **regions of interest (ROIs)** spanning the entire brain
- 114 **volumetric measurements** for larger structures calculated by combining single the single ROIs above **within a tree hierarchy**  

We check the **collinearity among all the imaging features** in the UPENN dataset and find: 
- (1) the 114 larger structures volumes are highly correlated with the single RIOs from which the larger structures volumes are calculated 
- (2) each pair of left and right ROIs (e.g., left hippocampus and right hippocampus) is also highly correlated. 

*(Figure 6)*

Therefore, we **restrict our analysis on the single ROIs** and keep the **average of each left and right ROI pair**, which results in **83 imaging features.**



Figure 6. **A Quick Look at the Correlations among Predictors**

```python
plt.figure(figsize=(20,20))
sb.set(font_scale=0.7)
sb.heatmap(abs(df_1101_aver_corr),cmap="YlGn",annot=False)
plt.title("A Quick Look at the Correlations among Predictors",fontsize=20);
#plt.yticks([5.8,4.8,4.3,3.1,2,0.4],corr_mx.columns,fontsize=7)
```
